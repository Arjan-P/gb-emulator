```python
bgMap=0x1800
colors={(0,0):' ',(0,1):'░',(1,0):'▒',(1,1):'█'}
```

```python

def scanLine(ly,scy):
    yPos=ly+scy
    tileRow=int(yPos/8)*32
    for pixel in range(160):
        xPos=pixel
        tileCol=int(xPos/8)
        tileIndex=vram[bgMap+tileRow+tileCol]
        tileData=vram[tileIndex*16:16+tileIndex*16]
        line=yPos%8
        line*=2
        b2=tileData[line]
        b1=tileData[line+1]
        colorBit=xPos%8
        colorBit-=7
        colorBit*=-1
        print(colors[((b1>>colorBit)&1,(b2>>colorBit)&1)],end='')
```
